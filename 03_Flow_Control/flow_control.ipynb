## Truth Values
The following values are considered falsey, in that they evaluate to False when applied to a boolean operator.
- None
- False
- 0, or any numerical value equivalent to zero, for example 0L, 0.0, 0j
- Empty sequences: '', "", (), []
- Empty mappings: {}
- User-defined types where the __bool__ or __len__ methods return 0 or False
- All other values in Python evaluate to True.
Note: A common mistake is to simply check for the Falseness of an operation which returns different Falsey values
where the difference matters. For example, using if foo() rather than the more explicit if foo() is None

## if else

syntax:
```
if condition:
    statement1
else:
    statement2
```

````python
# Write a program to check the input number is positive or negative.

number=int(input("Enter the number: "))
if number>0:
    print("Positive number")
else:
    print("Negative number")
````

````python
# example2 Write a Python program to check whether a number is even or odd.


number=int(input("Enter a number :" ))
if number%2==0:
    print("even number")
else:
    print("odd number")

# even number are easily divided by 2  and gives remainder 0 ex: 2,4,6,8,10--
# odd number are not easily divided by 2 and gives remainder 1 ex: 1,3,5,7,9--
````

````python
# Write a Python program to check whether a given string is a palindrome or not.

word=input("Enter a word :")
if word ==word[::-1]:          # checking if the string is equal to its reverse
    print(word ,"is palindrome")
else:
    print(word ,"is not palindrome")
````

# Nested if

Syntax
```
if condition:
    statement
    statement
    if condition:
        statement
        statement
    else:
        statement
else:
        statement
# rest of code
```


````python
# program for positive and negative numbers

number=float(input("Enter a number:"))
if number>=0:
    if number==0:
        print("Zero")
    else:
        print("The number is positive")
else:
    print("The Number is negative")
````

# if elif else

Syntax
```
if condition:
    statement
elif condition:
    statement
else:
    statement
# rest of code
```

````python
# example write a program 

number1=float(input("Enter a number:"))
number2=float(input("Enter a number:"))
operator=input("Enter your operator, +,-,*,/,% : ")
if operator =='+':
    print(number1+number2)
elif operator =='-':
    print(number1-number2)
elif operator =='*':
    print(number1*number2)
elif operator =='/':
    print(number1/number2)
elif operator =='%':
    print(number1-number2)
else:
    print("Enter valid operator")
````

````python
# garding

marks=float(input("Enter your marks:"))
if marks>85 and marks<=100:
    print("A grade")
elif marks>70 and marks<=85:
    print("B grade")
elif marks>55 and marks<=70:
    print("c grade")
elif marks>40 and marks<=55:
    print("D grade")
elif marks>30 and marks<=40:
    print("E grade")
else:
    print("Fail")

````

````python
# Write a Python program to find the factorial of a number.
num=int(input("Enter a number :"))
factorial=1
if num<0:
    print("No factorial")
elif num==0:
    print("factorial of 0 is 1")
else:
    for i in range(1,num+1):
        factorial=factorial*i
    print("factorial of",num,"is",factorial)
# factorial is the product of an integer and all the integers below ex 5 5!=5*4*3*2*1
````

Syntax:
```

    statement

```

````python
 x=1 
 while x <=10: 
    print(x) 
    x=x+1 
````

````python
#To display the sum of first n numbers
n=int(input("Enter number:")) 
sum=0 
i=1 
while i<=n: 
    sum=sum+i 
    i=i+1 
    print("The sum of first",n,"numbers is :",sum)
````


## Nested Loops:
Sometimes we can take a loop inside another loop,which are also known as nested loops.

````python
for i in range(4): 
    for j in range(4): 
        print("i=",i," j=",j) 
````

````python
# Q. Write a program to dispaly *'s in Right angled triangled form
n = int(input("Enter number of rows:")) 
for i in range(1,n+1): 
    for j in range(1,i+1): 
        print("*",end=" ") 
    print()
````

````python
n = int(input("Enter number of rows:")) 
for i in range(1,n+1): 
    print("* " * i) 
````

````python
# Q. Write a program to display *'s in pyramid style(also known as equivalent triangle)
n = int(input("Enter number of rows:")) 
for i in range(1,n+1): 
    print(" " * (n-i),end="") 
    print("* "*i)
````
